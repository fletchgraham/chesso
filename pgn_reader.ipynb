```python
import chess.pgn
```

```python
pieces = {
    'K' : 'King',
    'Q' : 'Queen',
    'B' : 'Bishop',
    'N' : 'Knight',
    'R' : 'Rook',
    'P' : 'Pawn',
}
```

```python
with open('lichess_pgn_2021.01.22_fletchgraham_vs_azza92f.HtYSCugn.pgn', 'r') as pgn:
    first_game = chess.pgn.read_game(pgn)
    
print(first_game)
```

```python
board = first_game.board()
board
```

```python
pause_duration = 1500

for move in first_game.mainline_moves():
    lan = board.lan(move)
    v = lan
    
    if lan == 'O-O':
        v = 'castles short'
    
    elif lan == 'O--O':
        v = 'castles long'
        
    else:
        v = v.replace('-', ' to ')
        v = v.replace('x', ' takes ')
        v = v.replace('+', ' check ')
        
        for code in pieces.keys():
            v = v.replace(code, f'{pieces.get(code)} ')
    
    print(f'{v} [[slnc {pause_duration}]]')
    board.push(move)
    

board
    
```
